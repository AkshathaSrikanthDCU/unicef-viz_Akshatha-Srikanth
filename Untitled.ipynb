```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the UNICEF deaths dataset
df = pd.read_csv("../unicef_indicator_1.csv")

# Load the world geojson file
world = gpd.read_file("countries.geojson")
```

```python
# Filter deaths aged 5 to 9
deaths = df[df['indicator'] == "Deaths aged 5 to 9"]

# Merge world map and death dataset
merged = world.merge(deaths, left_on='ISO3166-1-Alpha-3', right_on='alpha_3_code', how='left')

# Plot
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
merged.plot(
    column='obs_value',
    cmap='OrRd',
    linewidth=0.8,
    ax=ax,
    edgecolor='0.8',
    legend=True
)

# Beautify
plt.title('Deaths aged 5 to 9 (World Map)', fontsize=18)
plt.axis('off')
plt.show()
```
